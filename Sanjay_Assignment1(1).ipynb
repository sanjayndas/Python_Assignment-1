```python
!pip install --upgrade xlrd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# READING THE DATASET
data = pd.read_excel("/content/iris.xls")
print('Dataset of Iris Flower')
data.head()
```

Mean of Columns

```python

slmean = data['SL'].mean()
swmean = data['SW'].mean()
plmean = data['PL'].mean()
pwmean = data['PW'].mean()
mean = {'Data': ['SL','SW','PL','PW'],'Mean':[slmean,swmean,plmean,pwmean]}
mean_df = pd.DataFrame(mean,index = [1,2,3,4])
mean_df
```

Checking Null values in the Dataset

```python

print('No. of Null values in each Column:')
data.isnull().sum()
```

## Visualizations

1.Histogram


```python
plt.figure(figsize=(6,4))
plt.title('Count of Classification')
sns.countplot(x = data['Classification'])
plt.show()

plt.figure(figsize=(10,10))
plt.subplot(221)
plt.title('Petal Length (PL) (in cm)')
sns.histplot(x= data['PL'],color='green')

plt.subplot(222)
plt.title('Petal Width (PW) (in cm)')
sns.histplot(x= data['PW'])

plt.subplot(223)
plt.title('Sepal Length (SL) (in cm)')
sns.histplot(x= data['SL'],color='orange')

plt.subplot(224)
plt.title('Sepal Width (SW) (in cm)')
sns.histplot(x= data['SW'],color='red')
```

2.Scatter Plot

```python

plt.figure(figsize=(10,6))
plt.title('Relation between Petal Lenght(PL)(in cm) and Petal Width(PW)(in cm)')
sns.scatterplot(x= data['PL'],y = data['PW'], hue = data['Classification'])

plt.figure(figsize=(10,6))
plt.title('Relation between Sepal Lenght(SL)(in cm) and Sepal Width(SW)(in cm)')
sns.scatterplot(x= data['SL'],y = data['SW'], hue = data['Classification'])
```

3.Boxplot

```python

plt.figure(figsize=(10,10))
plt.subplot(221)
plt.title('Petal Length(PL)(in cm) variation')
sns.boxplot(x = data['Classification'], y = data['PL'])
plt.subplot(222)
plt.title('Petal Width(PW)(in cm) variation')
sns.boxplot(x = data['Classification'], y = data['PW'])
plt.subplot(223)
plt.title('Sepal Length(SL)(in cm) variation')
sns.boxplot(x = data['Classification'], y = data['SL'])
plt.subplot(224)
plt.title('Sepal Width(SW)(in cm) variation')
sns.boxplot(x = data['Classification'], y = data['SW'])
```

4.Violin Plot

```python

plt.figure(figsize=(10,10))
plt.subplot(221)
plt.title('Petal Length(PL)(in cm) variation')
sns.violinplot(y = data['PL'], x= data['Classification'])
plt.subplot(222)
plt.title('Petal Length(PL)(in cm) variation')
sns.violinplot(y = data['PW'], x= data['Classification'])
plt.subplot(223)
plt.title('Petal Length(PL)(in cm) variation')
sns.violinplot(y = data['SL'], x= data['Classification'])
plt.subplot(224)
plt.title('Petal Length(PL)(in cm) variation')
sns.violinplot(y = data['SW'], x= data['Classification'])
```

5.Pairplot

```python
sns.pairplot(data,hue='Classification')
```
